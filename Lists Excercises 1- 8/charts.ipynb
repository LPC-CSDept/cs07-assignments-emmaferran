```python
import matplotlib.pyplot as plt
import numpy as np

label = "Math", "Physics", "English", "Algebra"
x = np.arange(len(label))
y1 = [100, 90, 100, 90]
y2 = [100, 100, 100, 90]
width = .35

fig, ax = plt.subplots()
rect1 = ax.bar(x-width/2, y1, width, label="John")
rect2 = ax.bar(x+width/2, y2, width, label="Doe")

ax.set_ylabel("Scores")
ax.set_title("Scores by group and gender")
ax.legend()
plt.xticks(x, label)

ax.bar_label(rect1, padding = 0)
ax.bar_label(rect2, padding = 0)

plt.show()
```
