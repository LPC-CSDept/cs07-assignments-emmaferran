```python
#assignment 7.11
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
x = [1, 2, 3, 4]
Subjects = ["Math", "English", "Physics", "Computer"]
Bscores = [100, 90, 80, 60]
Mscores = [90, 80, 70, 100]



ax.bar(Subjects, Bscores, label= "Billy")
ax.bar(Subjects, Mscores, label= "Mary", bottom=Bscores)


ax.legend()
ax.set_title("Billy and Mary's Academic Performances")

plt.show()
```

```python
#assignment 7.12
import matplotlib.pyplot as plt
import numpy as np

label = ["Billy", "Mary"]
x = np.arange(2)
Math = [100, 90]
English = [90, 80]
Physics = [80, 70]
Computer = [60, 100]
width = .15

fig, ax = plt.subplots()
rect1 = ax.bar(x-width*1.5, Math, width, label= "Math")
rect2 = ax.bar(x-width*.5, English, width, label= "English")
rect3 = ax.bar(x+width*.5, Physics, width, label= "Physics")
rect4 = ax.bar(x+width*1.5, Computer, width, label= "Computer")

ax.set_title("Billy and Mary's Academic Scores")
ax.legend()
plt.xticks(x, label)

ax.bar_label(rect1, padding=3)
ax.bar_label(rect2, padding=3)
ax.bar_label(rect3, padding=3)
ax.bar_label(rect4, padding=3)

plt.show()
```
