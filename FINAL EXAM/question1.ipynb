```python
from functools import reduce

FLnames = open("fl2021.txt")
CAnames = open("fl2021.txt")

#pt1
CAinput = {
  "state": 'CA',
  "male": {
    "gender": 'Male',
    "name": [],
    "number": []
  },
  "female": {
    "gender": 'Female',
    "name": [],
    "number": []
  }
}
FLinput = {
  "state": 'FL',
  "male": {
    "gender": 'Male',
    "name": [],
    "number": []
  },
  "female": {
    "gender": 'Female',
    "name": [],
    "number": []
  }
}


def ReadFile(file, state):
    for line in file:
        line = line.strip()
        line = line.split(' ')
        state["male"]["name"].append(line[1])
        state["male"]["number"].append(line[2])
        state["female"]["name"].append(line[3])
        state["female"]["number"].append(line[4])

ReadFile(FLnames, FLinput)
ReadFile(CAnames, CAinput)

#pt2 
def GenderSort (genderlist, output, state):
    labels = ["state", "gender", "number", "name"]
    for num in range(len(gender["name"])):
        name = genderlist["name"][num]
        number = genderlist["number"][num]
        gender = genderlist("gender")
        data = [state]+[gender]+[number]+[name]
        output.append(dict(zip(labels, data)))
    return output

def DictionarySort (file):
    genderlist = []
    genderlist = GenderSort(file["male"], genderlist, file["state"])
    genderlist = GenderSort(file["female"], genderlist, file["state"])
    return genderlist

CAlist = DictionarySort(CAinput)
FLlist = DictionarySort(FLinput)
CAandFLlist = CAlist + FLlist

#pt3 and 4
def FindNames (data, state, gender, top, StartsWith, accumlative):
    data = list(filter(lambda x: x['state'] == state, data))
    data = list(filter(lambda x: x['gender'] == gender, data))
    if (StartsWith and StartsWith != ''): data = list(filter(lambda x: x["name"].startswith(StartsWith), data))
    data = sorted(data, key=lambda x: x["number"], reverse=True) 
    if (top and top != 0): data = data[:top] 
    if (accumlative == True): 
        data = list(map(lambda x:int(x["number"].replace(",","")), data)) 
    sumnum = lambda x,y: x+y
    data = reduce(sumnum, data, 0)
    return data

def PrintAll(list):
    for item in list:
        print(item)

PrintAll(FindNames(CAandFLlist, 'CA', 'Male', 10, '', False))
PrintAll(FindNames(CAandFLlist, 'FL', 'Female', 10, 'E', False))
print(FindNames(CAandFLlist, 'FL', 'Female', 10, '', True))
```
